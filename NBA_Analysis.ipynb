```python
import pandas as pd
```

```python
df_games = pd.read_csv("/content/games.csv")
df_details = pd.read_csv("/content/games_details.csv")
```

```python
df_games["GAME_ID"].value_counts()
```

```python
df_games = df_games.drop_duplicates(subset="GAME_ID")
```

```python
df_joined =  df_games.merge(df_details, on="GAME_ID", validate="one_to_many")
```

```python
df_joined
```

```python
df_joined = df_joined[df_joined["SEASON"] > 2018]
```

```python
df_joined["COMMENT"] = df_joined["COMMENT"].str[:3]
df_joined = df_joined[df_joined["COMMENT"].isnull()]
df_joined
```

```python
df_joined["AT_HOME"] = df_joined["TEAM_ID_home"] == df_joined["TEAM_ID"]
df_joined["PLAYER_WIN"] = df_joined["AT_HOME"] == df_joined["HOME_TEAM_WINS"]
df_joined["MONTH"] = df_joined["GAME_DATE_EST"].str[5:7].copy()
```

```python
df_joined["MONTH"] = df_joined["GAME_DATE_EST"].str[5:7].copy()
df_joined
```

```python
df_2021 = df_joined[df_joined["SEASON"] == 2021]
df_2021
```

```python

df_2021.columns
```

```python
df_2021["MIN"] = df_2021["MIN"].str[:2].copy()
df_2021['MIN'] = df_2021['MIN'].str.replace('\W', '', regex=True).astype(int)
df_2021 = df_2021[df_2021["MIN"] > 20]
```

```python
df_2021
```

```python
df_2021.groupby(["PLAYER_NAME", "GAME_ID"])[["PLAYER_WIN"]].mean().unstack("PLAYER_NAME").sum().unstack(level=0)
```

```python
pts_avg = df_2021.groupby(["PLAYER_NAME", "GAME_ID"])[["PTS"]].mean().unstack("PLAYER_NAME").mean().unstack(level=0)
fg3_pct_avg = df_2021.groupby(["PLAYER_NAME", "GAME_ID"])[["FG3_PCT"]].mean().unstack("PLAYER_NAME").mean().unstack(level=0)
fg_pct_avg = df_2021.groupby(["PLAYER_NAME", "GAME_ID"])[["FG_PCT"]].mean().unstack("PLAYER_NAME").mean().unstack(level=0)
ft_pct_avg = df_2021.groupby(["PLAYER_NAME", "GAME_ID"])[["FT_PCT"]].mean().unstack("PLAYER_NAME").mean().unstack(level=0)
stl_avg = df_2021.groupby(["PLAYER_NAME", "GAME_ID"])[["STL"]].mean().unstack("PLAYER_NAME").mean().unstack(level=0)
blk_avg = df_2021.groupby(["PLAYER_NAME", "GAME_ID"])[["BLK"]].mean().unstack("PLAYER_NAME").mean().unstack(level=0)
reb_avg = df_2021.groupby(["PLAYER_NAME", "GAME_ID"])[["REB"]].mean().unstack("PLAYER_NAME").mean().unstack(level=0)
to_avg = df_2021.groupby(["PLAYER_NAME", "GAME_ID"])[["TO"]].mean().unstack("PLAYER_NAME").mean().unstack(level=0)
plus_minus_avg = df_2021.groupby(["PLAYER_NAME", "GAME_ID"])[["PLUS_MINUS"]].mean().unstack("PLAYER_NAME").mean().unstack(level=0)
wins = df_2021.groupby(["PLAYER_NAME", "GAME_ID"])[["PLAYER_WIN"]].sum().unstack("PLAYER_NAME").sum().unstack(level=0)
```

```python
pts_avg
```

```python
df_season = pd.concat([
    pts_avg,
    fg3_pct_avg,
    fg_pct_avg,
    ft_pct_avg,
    stl_avg,
    blk_avg,
    reb_avg,
    to_avg,
    plus_minus_avg,
    wins
], axis=1)
```

```python
df_season.reset_index(inplace=True)
df_season
```

```python
df_2021 = df_season.merge(df_2021, on = "PLAYER_NAME",
                          suffixes = ("_avg", ""), validate="one_to_many")
```

```python
df_2021.columns
```

```python
df_2021[["TEAM_ID", "MONTH",
           "PLAYER_WIN"]] = df_2021[["TEAM_ID",
                                       "MONTH",
                                       "PLAYER_WIN"]]
```

```python
X_train = df_2021[["MONTH", "TEAM_ABBREVIATION", 'PTS_avg', 'FG_PCT_avg',
                   'FT_PCT_avg', 'STL_avg', 'BLK_avg', 'REB_avg', 'TO_avg',
                   'PLUS_MINUS_avg', 'PLAYER_WIN_avg']]
y_train = df_2021["PLAYER_WIN"]
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, mean_squared_error

# Load the data into a pandas DataFrame
data = df_2021

# Preprocess the data
# Convert 'TEAM_NAME' to binary columns using one-hot encoding
enc = OneHotEncoder()
team_name_enc = enc.fit_transform(data[['TEAM_ABBREVIATION']]).toarray()
team_name_cols = enc.get_feature_names_out(['TEAM_ABBREVIATION'])
team_name_df = pd.DataFrame(team_name_enc, columns=team_name_cols)

# Concatenate the binary columns and numerical columns
X = pd.concat([team_name_df, data[['MONTH', 'PTS_avg', 'FG_PCT_avg', 'FT_PCT_avg', 'STL_avg', 'BLK_avg',
                                   'REB_avg', 'TO_avg', 'PLUS_MINUS_avg']]], axis=1)
y = data['PLAYER_WIN']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)

```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)
model = MLPClassifier(solver = 'sgd', hidden_layer_sizes= (16, 8, 4),
                      alpha= 0.001, activation= 'relu')
model.fit(X_train, y_train)

# Evaluate the model on the testing data
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
print("Accuracy: {:.2f}".format(accuracy))
print("Precision: {:.2f}".format(precision))
print("Recall: {:.2f}".format(recall))
print("F1 score: {:.2f}".format(f1))
```

```python
def create_lineup(abr):
  df = df_2021[df_2021["TEAM_ABBREVIATION"] == abr].copy()
  df = df.iloc[:,0:11].drop_duplicates()
  final_lineup = df.sort_values(by="PLAYER_WIN_avg", ascending=False)[0:6]
  final_lineup["MONTH"] = 4
  final_lineup["TEAM_ABBREVIATION"] = abr
  return final_lineup
```

```python
def predict_result(player):
  X = df_season[df_season["PLAYER_NAME"] == player].copy()
  X["MONTH"] = 4

  columns_to_suffix = ['PTS', 'FG3_PCT', 'FG_PCT', 'FT_PCT',
                       'STL', 'BLK', 'REB', 'TO',
                       'PLUS_MINUS','PLAYER_WIN']
  X = X.iloc[:, 1:]
  suffix = '_avg'
  for col in columns_to_suffix:
    if col in X.columns:
        X.rename(columns={col: col + suffix}, inplace=True)

  prob=pipeline.predict_proba(X)[:,1]

  return prob
```

```python
import numpy as np

teams = df_2021["TEAM_ABBREVIATION"].drop_duplicates().values

team_results = []

for team in teams:
  lineup = create_lineup(team)
  player_results = []

  for player in lineup["PLAYER_NAME"]:
    player_results.append(predict_result(player))
  team_results.append(sum(player_results)/len(player_results))

results = np.squeeze(team_results)

data = {'Team': teams, 'Probability': results}

df = pd.DataFrame(data).sort_values(by="Probability", ascending=False)

print(df)
```
